# Importi

```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import imageio
```

# Funkcije

```python
def ucitaj_snimak(brojSnimka):
    return imageio.get_reader('snimci/video'+str(brojSnimka)+'.mp4','mp4')
```

# Izdvajanje Platoa

```python
snimak = ucitaj_snimak(1)
slika = snimak.get_data(0)

invertovana_slika = 255-slika

slika_siva = cv2.cvtColor(invertovana_slika, cv2.COLOR_RGB2GRAY)
ret, binarna = cv2.threshold(slika_siva, 200, 255, cv2.THRESH_BINARY)

kernel = np.ones((3, 3))
binarna = cv2.erode(binarna, kernel, iterations=1)
    
image, contours, hierarchy = cv2.findContours(binarna, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

slika_konture = slika.copy()
cv2.drawContours(slika_konture, contours, -1, (0, 255, 0), 1)
#plt.imshow(slika_konture)
#plt.show()

plato = max(contours, key=cv2.contourArea) #kontura koja obuhvata plato je najveca i nju izdvajamo

izdvojen_plato = slika.copy()
cv2.drawContours(izdvojen_plato, plato, -1, (0, 255, 0), 1)
plt.imshow(izdvojen_plato)
plt.show()
```
